```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
from utils import load_all_data, normalise_data
```

```python
# First examine the data

X_train, Y_train, time_train, X_test, Y_test, time_test = load_all_data()

fig, axs = plt.subplots(2, 1, figsize=(10, 10))
for i in range(len(X_train)):
    axs[0].plot(time_train, Y_train[i], alpha=0.5, c='b')
    axs[0].set_title('Training data')
for i in range(len(X_test)):
    axs[1].plot(time_test, Y_test[i], alpha=0.5, c='b')
    axs[1].set_title('Test data')
plt.show()
```

```python
# Reduce dimensionality of the feature data to 2D using PCA
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].scatter(X_train_pca[:, 0], X_train_pca[:, 1], c='b')
axs[0].set_title('Training data')
axs[1].scatter(X_test_pca[:, 0], X_test_pca[:, 1], c='b')
axs[1].set_title('Test data')
plt.show()
```
